```python
import matplotlib.pyplot as plt
import numpy as np

import fnmatch
import os
import pickle

#import yasa
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(font_scale=1.2)
import mne
import numpy as np

#e = mne.io.read_raw('/Volumes/Research/eng_research_scottlab/RATACAD_DATA/rawEEG/2022-06-08_raw.fif.gz',on_split_missing ='ignore')
e = mne.io.read_raw('/projectnb/ratacad/qdo/rawEEG/2022-03-18_raw.fif.gz',on_split_missing ='ignore')
steps = mne.find_stim_steps(e,stim_channel=['STIM3'])
itemindex = np.where(steps[:,2] == 4)
timetrial = steps[itemindex]

plt.hist(np.floor(timetrial[:,0]/(1000*3600)),bins=24)
plt.show()
#getting Choice Event aligned
events = mne.find_events(e, stim_channel=['STIM3'],shortest_event=1,min_duration=0.001,initial_event=True,consecutive=True)
print(events[:6])  # show the first 5
event_dict = {'choice':4,
             'stim':3,
             'center':2,
             'nothing':1}
fig = mne.viz.plot_events(events, event_id=event_dict, sfreq=e.info['sfreq'],
                          first_samp=e.first_samp)
#align by Choice
choiceLabel = np.array(dt.loc[:,"choice"])
rewardLabel = np.array(dt.loc[:,"reward"])
ChoiceEvent=events[events[:,2]==4]
ChoiceEvent.shape
```

```python
for i in range(len(choiceLabel)):
    if (choiceLabel[i] == "left") and (rewardLabel[i] == 0):
        choiceLabel[i] = 0 #left wrong
    elif (choiceLabel[i] == "left") and (rewardLabel[i] == 1):
        choiceLabel[i] = 1 #Left correct
    elif (choiceLabel[i] == "right") and (rewardLabel[i] == 0):
        choiceLabel[i] = 2 #right wrong
    elif (choiceLabel[i] == "right") and (rewardLabel[i] == 1):
        choiceLabel[i] = 3 #right correct
    elif choiceLabel[i] == "earlyleft":
        choiceLabel[i] = 4 #early left
    else:
        choiceLabel[i] = 5 #early right
ChoiceEvent[:,2] = choiceLabel
event_dict = {'left_wrong':0,
             'left_right':1,
             'right_wrong':2,
             'right_right':3,
             'early_left':4,
             'early_right':5}
fig = mne.viz.plot_events(ChoiceEvent, event_id=event_dict, sfreq=e.info['sfreq'],
                          first_samp=e.first_samp)
#align by choice
choice_epochs = mne.Epochs(e, ChoiceEvent,event_id=event_dict, tmin=-3, tmax=1, preload=True,event_repeated='drop')
conds_we_care_about = ['right_right','left_right']
choice_epochs.equalize_event_counts(conds_we_care_about)  # this operates in-place

choice_epochs['right_right'].plot_image(picks=['EEG001'])
choice_epochs['left_right'].plot_image(picks=['EEG001'])
choice_epochs['right_right'].plot_image(picks=['EMG001'])
choice_epochs['left_right'].plot_image(picks=['EMG001'])
#epochs['eleft'].plot_image(picks=['EEG001'])
#epochs['eright'].plot_image(picks=['EEG001'])
#frequencies = np.arange(7, 80, 3)


```

```python
frequencies = np.arange(1., 100., 2)  # define frequencies of interest
n_cycles = frequencies / 5.  # different number of cycle per frequency
power = mne.time_frequency.tfr_morlet(choice_epochs['left_wrong'],  n_cycles=n_cycles, return_itc=False,
                                      freqs=frequencies, decim=3,use_fft=True)
power.plot(['EEG001'],mode='ratio',baseline=(-3, -2),dB=True,vmin=-2,vmax=2)

power = mne.time_frequency.tfr_morlet(choice_epochs['right_wrong'], n_cycles=n_cycles, return_itc=False,
                                      freqs=frequencies, decim=3,use_fft=True)
power.plot(['EEG001'],mode='ratio',baseline=(-3, -2),dB=True,vmin=-2,vmax=2)
```

```python
#align by Initiation of Trial
events = mne.find_events(e, stim_channel=['STIM3'],shortest_event=1,min_duration=0.001,initial_event=True,consecutive=True)
print(events[:6])  # show the first 5
event_dict = {'choice':4,
             'stim':3,
             'center':2,
             'nothing':1}
#exclude 1 and 3 from event marking, and keep only 2 (center) and 4 (poke) 
#align by center in
event_simple = mne.pick_events(events, exclude=[1,3])
event_simple.shape
savedEvent = []
#align to start of Trial!
for i in range(event_simple.shape[0]):
    if event_simple[i,2] == 2:
        if (i+1) < event_simple.shape[0]:
            if event_simple[i+1,2] == 4:
                savedEvent.append(event_simple[i,:])
            else:
                pass
        
print(len(savedEvent))
savedEvent = np.array(savedEvent)
savedEvent.shape
choiceLabel = np.array(dt.loc[:,"choice"])
rewardLabel = np.array(dt.loc[:,"reward"])
for i in range(len(choiceLabel)):
    if (choiceLabel[i] == "left") and (rewardLabel[i] == 0):
        choiceLabel[i] = 0 #left wrong
    elif (choiceLabel[i] == "left") and (rewardLabel[i] == 1):
        choiceLabel[i] = 1 #Left correct
    elif (choiceLabel[i] == "right") and (rewardLabel[i] == 0):
        choiceLabel[i] = 2 #right wrong
    elif (choiceLabel[i] == "right") and (rewardLabel[i] == 1):
        choiceLabel[i] = 3 #right correct
    elif choiceLabel[i] == "earlyleft":
        choiceLabel[i] = 4 #early left
    else:
        choiceLabel[i] = 5 #early right
savedEvent[:,2] = choiceLabel
event_dict = {'left_wrong':0,
             'left_right':1,
             'right_wrong':2,
             'right_right':3,
             'early_left':4,
             'early_right':5}
fig = mne.viz.plot_events(savedEvent, event_id=event_dict, sfreq=e.info['sfreq'],
                          first_samp=e.first_samp)

epochs = mne.Epochs(e, savedEvent,event_id=event_dict, tmin=-0.5, tmax=2, preload=True,event_repeated='drop')
conds_we_care_about = ['left_right','right_right']
epochs.equalize_event_counts(conds_we_care_about)  # this operates in-place

epochs['left_right'].plot_image(picks=['EEG001'])
epochs['right_right'].plot_image(picks=['EEG001'])


```

```python
frequencies = np.arange(1., 80., 2)  # define frequencies of interest
n_cycles = frequencies / 5.  # different number of cycle per frequency
power = mne.time_frequency.tfr_morlet(epochs['right_right'],  n_cycles=n_cycles, return_itc=False,
                                      freqs=frequencies, decim=3,use_fft=True)
power.plot(['EEG001'],mode='ratio',baseline=(-0.5, 0.1),dB=True,vmin=-2,vmax=2)

power = mne.time_frequency.tfr_morlet(epochs['right_wrong'], n_cycles=n_cycles, return_itc=False,
                                      freqs=frequencies, decim=3,use_fft=True)
power.plot(['EEG001'],mode='ratio',baseline=(-0.5, 0.1),dB=True,vmin=-2,vmax=2)
```
